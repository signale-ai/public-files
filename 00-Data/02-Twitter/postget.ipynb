```python
# follow setup instructions in postget Github repo: https://github.com/alessandriniluca/postget
# postget works by login into your Twitter account and then scrapping the html code
# as of 2nd Jul 2023, you can read 10,000 tweets for blue check verified accounts, 1000 tweets for unverified accounts and 500 tweets for new unverified accounts
from postget.Posts import Posts
from postget.exceptions.exceptions import *
import pandas as pd

def main():
    twitter_getter = Posts(username='MY-TWITTER-USERNAME', 
                           password='MY-TWITTER-PASSWORD', 
                           num_scrolls=10, 
                           query='', # Note: set keyword query below
                           email_address='MY-TWITTER-EMAIL', 
                           mode=1, 
                           until='2023-01-14', 
                           since='2023-01-01', 
                           headless=False,  # need to set to False
                           chromedriver='C:/Users/leo_c/Downloads/chromedriver_win32'
                           )
    try:
        twitter_getter.login()
    except ElementNotLoaded as e:
        raise e


    print("Setting query in the object")
    twitter_getter.set_query('(crypto OR cryptocurrency)') # Note: set keyword query here

    print("Start Search, this will input the query and perform the scroll with the selected mode")
    try:
	    twitter_getter.search()
    except ElementNotLoaded as e:
	    raise e
    except NoTweetsReturned as e:
	    print(e)
            
    # Create dataframe of tweets
    df_tweets = pd.DataFrame.from_dict(twitter_getter.tweets, orient='index')
	
    print("Printing returned results and going home")
    twitter_getter.print_results()
    twitter_getter.go_home()

    print("Clearing Results")
    twitter_getter.clear_tweets()

    print("quitting browser")
    twitter_getter.quit_browser()

    return df_tweets

if __name__=='__main__':
    df = main()
    
df
```
