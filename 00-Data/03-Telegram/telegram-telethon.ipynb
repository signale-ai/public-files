```python
import asyncio
import nest_asyncio
from telethon.sync import TelegramClient
import pandas as pd

# Follow the steps in the links below to obtain the API ID and API hash for your Telegram account:
# - https://python.plainenglish.io/how-to-scrape-everything-from-telegram-using-python-f71437194569
# - https://stackoverflow.com/questions/38104560/telegram-api-create-new-application-error

username = "MY-USERNAME"   # your Telegram account username
api_id = MY-API-ID   # your Telegram account API ID
api_hash = "MY-API-HASH"   # your Telegram account API hash
phone = "MY-MOB-NO"  # your Telegram account mob. no. with country code (eg. +61481842767). You will be prompted to enter this to get a verification code on your phone
channel_username = "lobsters_chat"  # channel username
start_date_value = "2023-06-01 00:00:00"  # Specify the date and time range (in UTC)
end_date_value = "2023-07-25 23:59:59"    # Specify the date and time range (in UTC)
keywords_value = [] # Specify the keywords to filter, eg. keywords_value = ["Ethereum", "Eth", "Ether"]. If no keywords given, then no keyword filter is applied

# Apply nest_asyncio to enable running an event loop within a running loop in Jupyter Notebook
nest_asyncio.apply()  

async def main(start_date=None, end_date=None, keywords=None):
    # Convert start date and time range to timestamp format
    start_date = pd.Timestamp(start_date).timestamp()

    # Convert keywords to lowercase
    keywords = [keyword.lower() for keyword in keywords]
    
    data = [] 
    async with TelegramClient(username, api_id, api_hash) as client:
        # Offset date is retrieving messages previous to this date (exclusive)
        async for message in client.iter_messages(entity = "https://t.me/" + channel_username, offset_date = pd.Timestamp(end_date) + pd.Timedelta(seconds=1)):
            if start_date <= pd.Timestamp(message.date).timestamp():
                # Convert message to lowercase and split the message into individual words
                words_lower_in_msg = str(message.text).lower().split()
                # 'if not keywords' means when no keywords are given, ie. keywords_value = []
                if not keywords or any(keyword == word_lower_in_msg for word_lower_in_msg in words_lower_in_msg for keyword in keywords):
                    print(message.date, message.text, message.sender_id) 
                    data.append([message.date, message.text, message.sender_id])
            # To break the async loop when the date of the message is older than start date, ie. when start_date > pd.Timestamp(message.date).timestamp()
            else:
                break

    # Creates a new dataframe
    df = pd.DataFrame(data, columns=["message.date", "message.text", "message.sender_id"])

    # Creates a csv file
    df.to_csv('filename.csv', encoding='utf-8')

    return df

# Get the event loop and run the main function
loop = asyncio.get_event_loop()
df = loop.run_until_complete(main(start_date = start_date_value, end_date = end_date_value, keywords = keywords_value))
df
```
